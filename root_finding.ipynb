# Теоретическая справка: Локализация корней, метод касательных и метод хорд

## 1. Локализация корней уравнения $f(x) = 0$

**Определение**:  
Процесс нахождения отрезков $[a, b]$, на которых уравнение $f(x) = 0$ имеет единственный корень.

**Условия применения**:  
1. Функция $f(x)$ непрерывна на $[a, b]$ (теорема Больцано-Коши).  
2. На концах отрезка значения функции имеют разные знаки:  
   $$ f(a) \cdot f(b) < 0 $$  
3. Производная $f'(x)$ сохраняет знак на $[a, b]$ (гарантия единственности корня).

**Графический метод**:  
Приведение уравнения к виду $\phi(x) = \psi(x)$ и анализ точек пересечения графиков.

## 2. Метод касательных (Ньютона)

**Идея метода**:  
Аппроксимация функции касательной в начальной точке $x_0$ с последующим нахождением точки пересечения с осью $OX$.

**Формула итерации**:  
$$ x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}, \quad n = 0, 1, 2, \dots $$

**Условия сходимости**:  
1. $f(x)$ дважды дифференцируема на $[a, b]$.  
2. $f'(x) \neq 0$ на $[a, b]$.  
3. $f''(x)$ сохраняет знак на $[a, b]$.  
4. Начальное приближение $x_0$ выбирается так, что:  
   $$ f(x_0) \cdot f''(x_0) > 0 $$

**Геометрическая интерпретация**:  
![Метод Ньютона](https://upload.wikimedia.org/wikipedia/commons/e/e0/NewtonIteration_Ani.gif)

## 3. Метод хорд

**Идея метода**:  
Аппроксимация функции хордой, соединяющей точки $(a, f(a))$ и $(b, f(b))$, с нахождением точки пересечения с осью $OX$.

**Формула итерации**:  
- Неподвижный левый конец ($f(a) \cdot f''(a) > 0$):  
  $$ x_{n+1} = x_n - \frac{f(x_n) \cdot (a - x_n)}{f(a) - f(x_n)} $$  
- Неподвижный правый конец ($f(b) \cdot f''(b) > 0$):  
  $$ x_{n+1} = x_n - \frac{f(x_n) \cdot (b - x_n)}{f(b) - f(x_n)} $$

**Условия сходимости**:  
1. $f(x)$ непрерывна на $[a, b]$.  
2. $f(a) \cdot f(b) < 0$.  
3. $f'(x)$ и $f''(x)$ сохраняют знак на $[a, b]$.

**Сравнение методов**:  
| Критерий          | Метод Ньютона           | Метод хорд               |
|-------------------|-------------------------|--------------------------|
| Скорость сходимости | Квадратичная ($O(n^2)$) | Линейная ($O(n)$)        |
| Требования        | Нужна $f'(x)$           | Только $f(x)$            |
| Устойчивость      | Чувствителен к $x_0$    | Более устойчив           |

## 4. Комбинированный метод (метод хорд и касательных)

**Применение**:  
Используется для ускорения сходимости:  
- Метод Ньютона работает с "выпуклой" стороны.  
- Метод хорд — с "вогнутой".

1. **Метод касательных (Ньютона)**:
   $$\bar{x}_{n} = \bar{x}_{n-1} - \frac{f(\bar{x}_{n-1})}{f'(\bar{x}_{n-1})},$$
   где:
   - $\bar{x}_{n}$ — очередное приближение по методу касательных,
   - $f'(\bar{x}_{n-1})$ — значение производной в точке $\bar{x}_{n-1}$.

2. **Метод хорд**:
   $$x_{n} = x_{n-1} - \frac{f(x_{n-1}) \cdot (\bar{x}_{n-1} - x_{n-1})}{f(\bar{x}_{n-1}) - f(x_{n-1})},$$
   где:
   - $x_{n}$ — очередное приближение по методу хорд,
   - $\bar{x}_{n-1} - x_{n-1}$ — длина текущего интервала.

3. **Условие остановки**:
   $$|x_{n} - \bar{x}_{n}| < \varepsilon,$$

## Обозначения
- $\bar{x}_{0}, \bar{x}_{1}, \ldots$ — последовательность приближений по методу касательных,
- $x_{0}, x_{1}, \ldots$ — последовательность приближений по методу хорд.


## Анализ функции $f(x) = 2\ln x - \frac{1}{x}$

### Область определения:
Функция определена только при $x > 0$, так как:
- $\ln x$ существует при $x > 0$
- Знаменатель $\frac{1}{x}$ требует $x \neq 0$

Поэтому все дальнейшие производные и их свойства рассматриваются исключительно на интервале $x > 0$.

### 1. Первая производная:
$$
f'(x) = \frac{2}{x} + \frac{1}{x^2} = \frac{2x + 1}{x^2} > 0 \quad \forall x > 0
$$
**Вывод:** Функция строго возрастает на $(0, +\infty)$ → корень единственный.

### 2. Вторая производная:
$$
f''(x) = -\frac{2}{x^2} - \frac{2}{x^3} = -\frac{2(x + 1)}{x^3} < 0 \quad \forall x > 0
$$
**Следствие:** Функция выпукла вверх на всей области.

### 3. Локализация корня:
| $x$  | $f(x)$          |
|------|-----------------|
| $1$  | $-1$            |
| $2$  | $\approx 0.886$ |

**Итог:** Единственный корень $\in (1, 2)$.

### Критерии единственности:
1. $f'(x) > 0$ → монотонность
2. $f(a)\cdot f(b) < 0$ → существование
3. $f''(x)$ сохраняет знак → отсутствие перегибов

```python
def find_intervals(f: callable, 
                   start: float, 
                   end: float, 
                   step: float = 1
                   ) -> tuple:
    """
    Finds intervals where the function changes sign (potential roots).
    
    Args:
        f: Function to analyze.
        start: Start of the search range.
        end: End of the search range.
        step: Step size for creating the search grid (default 1).
    
    Returns:
        A tuple of intervals (as tuples) where the function changes sign.
    """
    grid = np.arange(start, end, step)
    intervals = []
    for i in range(len(grid) - 1):
        if f(grid[i]) * f(grid[i + 1]) <= 0:
            intervals.append((grid[i], grid[i + 1]))
    return intervals
```

```python
def get_point_for_newtons_method(f: callable, f_2: callable, start: float, end: float) -> float:
    """
    Selects the appropriate starting point for Newton's method.
    
    Args:
        f: The target function.
        f_2: The second derivative of the function.
        start: Left endpoint of the interval.
        end: Right endpoint of the interval.
    
    Returns:
        The optimal starting point (start or end) for Newton's method.
    """
    if f(start) * f_2(start) < 0:
        return start
    else:
        return end
```

```python
def newtons_method(f: callable, 
                   f_1: callable, 
                   x_0: float, 
                   print_results: bool = False, 
                   eps: float = 1e-7
                   ) -> float:
    """
    Finds a root of the function using Newton's method.
    
    Args:
        f: The target function.
        f_1: The first derivative of the function.
        x_0: Initial approximation.
        print_results: Whether to print iteration details (default False).
        eps: Tolerance for convergence (default 1e-7).
    
    Returns:
        The approximated root of the function.
    """
    x = x_0
    x_prev = x + eps + 1
    errors = []
    
    if print_results:
        print("Root approximations:")
        
    i = 0
    while abs(x_prev - x) > eps:
        x_new = x - f(x) / f_1(x)
        x_prev = x
        x = x_new
        if print_results:
            i += 1
            errors.append(abs(x - x_prev))
            print(f"{i}: {x}, {x_prev}")
    
    if print_results:
        print(f"Total iterations: {i}")
        plt.figure(figsize=(6, 3))
        plt.plot(range(1, len(errors) + 1), errors, 'r-')
        plt.title('Error dependence on iteration number for Newton\'s method')
        plt.xlabel('Iteration')
        plt.ylabel('Error')
        plt.grid(True)
        plt.show()
    
    return x
```

# Пример использования функции

```python
y = lambda x: 2 * np.log(x) - 1 / x
y_1 = lambda x: 2 / x + 1 / x ** 2
y_2 = lambda x: -2 / x ** 2 - 2 / x ** 2

a = 0.1
b = 100

intervals = find_intervals(y, a, b, 3)
root_quantity = len(intervals)

x_0 = get_point_for_newtons_method(y, y_2, intervals[0][0], intervals[0][1])
root = newtons_method(f=y, f_1=y_1, x_0=x_0, print_results=True)

print(f"Root (precision 10^-7): {root}, function value: {y(root)}")
```

```python
def get_points_for_chord_method(
        f: callable, 
        f_2: callable, 
        start: float, 
        end: float
) -> tuple[float, float]:
    """
    Selects appropriate starting points for the chord method.
    
    Args:
        f: The target function.
        f_2: The second derivative of the function.
        start: Left endpoint of the interval.
        end: Right endpoint of the interval.
    
    Returns:
        A tuple of optimal starting points (a, b) for the chord method.
    """
    if f(start) * f_2(start) > 0:
        return (start, end)
    else:
        return (end, start)
```

```python
def chord_method(f: callable, 
                 a: float, 
                 b: float,
                 print_results: bool = False, 
                 eps: float = 1e-7
                 ) -> float:
    """
    Finds a root of the function using the chord method.
    
    Args:
        f: The target function.
        a: Left endpoint of the interval.
        b: Right endpoint of the interval.
        print_results: Whether to print iteration details (default False).
        eps: Tolerance for convergence (default 1e-7).
    
    Returns:
        The approximated root of the function.
    
    Raises:
        ValueError: If there's no root in the given interval.
    """
    if f(a) * f(b) > 0:
        raise ValueError("There is no root in this segment")
    
    x = a
    x_prev = b
    errors = []
    
    if print_results:
        print("Root approximations:")
        
    i = 0
    while abs(x_prev - x) > eps:
        x_new = x - f(x) * (b - x) / (f(b) - f(x))
        x_prev = x
        x = x_new
        if print_results:
            i += 1
            print(f"{i}: {x}, {x_prev}")
            errors.append(abs(x - x_prev))
    
    if print_results:
        print(f"Total iterations: {i}")
        plt.figure(figsize=(6, 3))
        plt.plot(range(1, len(errors) + 1), errors, 'b-')
        plt.title('Error dependence on iteration number for Chord method')
        plt.xlabel('Iteration')
        plt.ylabel('Error')
        plt.grid(True)
        plt.show()
    
    return x
```

```python
y = lambda x: 2 * np.log(x) - 1 / x
y_1 = lambda x: 2 / x + 1 / x ** 2
y_2 = lambda x: -2 / x ** 2 - 2 / x ** 2

a = 1
b = 100

intervals = find_intervals(y, a, b, 3)
root_quantity = len(intervals)

start, end = get_points_for_chord_method(y, y_2, intervals[0][0], intervals[0][1])
root = chord_method(f=y, a=start, b=end, print_results=True)

print(f"Root (precision 10^-7): {root}, function value: {y(root)}")
```
